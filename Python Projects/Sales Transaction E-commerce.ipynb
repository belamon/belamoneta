```python
## Questions

- what is the most frequently purchased products 
- most profitable customer segment 
- sales over the months 
- canceled products 
- recommendation for increasing profit 
```

```python
# what is the most frequently purchased products 
# step 1 : group and aggregate data
# step 2 : sort the product 
# step 3 : display top 10 product 

top_purchased_products =  df.groupby(['ProductNo','ProductName'])['Quantity'].sum().reset_index().sort_values(by='Quantity', ascending=False).head(10)
top_purchased_products 
```

```python
#most profitable costumer segment by county 
#make new column called total 
df['Total']=df['Price'] * df['Quantity']
#find the most profitable costumer segment
most_profitable_segment = df.groupby('Country')['Total'].sum().reset_index().sort_values(by='Total', ascending=False).head(10)
most_profitable_segment 
```

```python
#find sales over the months 
# since we have converted the date coloumn to datetime format, now make a new coloumn
# name the new coloumn as month 
df['MonthYear']= df['Date'].dt.to_period('M')
monthly_sales= df.groupby('MonthYear')['Total'].sum().reset_index()
monthly_sales
```

```python
#find canceled products 
total_canceled_products =df[df['Quantity'] <0].groupby('ProductName')['Total'].sum().reset_index().sort_values(by='Total', ascending=True)
total_canceled_products
```

```python
from sc_google_maps_api import ScrapeitCloudClient
import json
import pandas as pd 

api_key = 'e505c973-d5eb-43df-8ecc-0f796bea8773'
client = ScrapeitCloudClient(api_key=api_key)

# Define the search query and parameters
search_query = "Restaurants in South Jakarta"
country = "ID"
domain = "com"
num_pages = 5

# Loop through the pages
for page in range(1, num_pages + 1):
    params = {
        "keyword": search_query,
        "country": country,
        "domain": domain,
        "page": page  # Specify the page number
    }
    
    response = client.scrape(params=params)

    # Check if the request was successful
    if response.status_code == 200:
        # Print the response text for the current page
        print(f"Page {page} Results:")
        print(response.text)
    else:
        print(f"Failed to retrieve data from page {page}")

```

```python
 # Parse the JSON response
data = json.loads(response.text)

    # Extract relevant information into separate lists
titles = [hotel["title"] for hotel in data["scrapingResult"]["locals"]]
phones = [hotel["phone"] for hotel in data["scrapingResult"]["locals"]]
addresses = [hotel["address"] for hotel in data["scrapingResult"]["locals"]]
ratings = [hotel["rating"] for hotel in data ["scrapingResult"]["locals"]]
websites = [hotel["website"]for hotel in data["scrapingResult"]["locals"]]

    # Create a DataFrame
df = pd.DataFrame({
        "Title": titles,
        "Phone": phones,
        "Address": addresses,
        "Rating":ratings,
        "Website":websites
    })
```
